```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split

import os, sys
```

```python
cwd = os.getcwd()
train_data_path = os.path.join(cwd, 'data', 'train.csv')
test_data_path = os.path.join(cwd, 'data', 'test.csv')

train_data = pd.read_csv(train_data_path)
test_data = pd.read_csv(test_data_path)
```

```python
# first 25 columns are features, last column is target in train data
# use train test split

X = train_data.iloc[:, :-1]
y = train_data.iloc[:, -1]

# normalize data
def normalize(data):
    return (data - data.mean()) / data.std()

def preprocess_data(data: pd.DataFrame):
    # handle normalisation
    data = normalize(data)
    # convert to numpy array
    return np.array(data)

X = preprocess_data(X)
test_data = preprocess_data(test_data)

# delete ID column in test data
test_data = np.delete(test_data, 0, axis=1)
```

```python
# define model without jax using linear regression

def linear_regression(X, y, alpha=0.01, epochs=1000):
    # add bias column
    X = np.hstack((np.ones((X.shape[0], 1)), X))
    # initialize weights
    theta = np.zeros(X.shape[1])
    # gradient descent
    for _ in range(epochs):
        theta = theta - (alpha / X.shape[0]) * np.sum((X @ theta - y) * X.T, axis=1)
    return theta

# train model
theta = linear_regression(X, y)

# plot loss
def plot_loss(X, y, theta):
    X = np.hstack((np.ones((X.shape[0], 1)), X))
    loss = np.sum((X @ theta - y) ** 2) / (2 * X.shape[0])
    return loss

loss = plot_loss(X, y, theta)
print(loss)





```
